```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel("sales_data.xlsx")
df=pd.DataFrame(data)
ds=df.groupby("Payment Mode")["Price"].sum()
plt.figure(facecolor="silver")

bars=ds.plot(kind="bar", color=["red", "blue", "orange"],edgecolor="black" )
plt.gca().set_facecolor("lightpink") 
for bar in bars.patches:
    bars.text(bar.get_x() + bar.get_width()/2,
              bar.get_height(),
              int(bar.get_height()),
              ha='center', va='bottom')
plt.title("Expense Tracker Visualization",fontsize=26, fontweight="bold", color="blue")
plt.xlabel("Payment Mode",fontsize=18)
plt.ylabel("Amount",fontsize=18)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.savefig("Expense_tracker.png")
plt.show()

```

```python
ms=df.groupby("Payment Mode").agg({"Product": lambda x: ', '.join(x), "Price":"sum"}).reset_index()
ms.columns = ["Payment Mode", "Product","Price"]
ms.to_csv("Expense Tracker.csv", index=False)
ms.to_excel("Expense_Tracker.xlsx", index=False)
```
